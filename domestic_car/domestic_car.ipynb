```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# ChromeDriver 초기화
driver = webdriver.Chrome()

# 대상 URL 접속
url = 'https://auto.danawa.com/auto/?Work=record'
driver.get(url)

# 페이지 로드 대기
time.sleep(3)

# 국산차 브랜드 리스트
domestic_brands = ['현대', '기아', '르노코리아', '쌍용', '제네시스', '쉐보레']

# 데이터 저장 리스트
data = []

try:
    for year in range(2023, 2026):  # 연도 범위 설정
        for month in range(1, 13):  # 월 범위 설정
            if year == 25 and month > 2:  # 현재 날짜 기준으로 데이터 제한
            break
            
            print(f"📌 {year}년 {month}월 데이터 수집 시작...")

            # 기간 설정 (라디오 버튼 클릭)
            radio_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="monthPeriodDiv"]/span[2]/label/input'))
            )
            radio_button.click()
            time.sleep(1)  # 값이 적용될 시간 대기

            # 시작 연도 & 월 설정
            from_year = Select(driver.find_element(By.XPATH, '//*[@id="selMonthFrom"]'))
            from_year.select_by_value(str(year))
            time.sleep(0.5)  # 반영 대기

            from_month = Select(driver.find_element(By.XPATH, '//*[@id="selDayFrom"]'))
            from_month.select_by_value(f"{month:02d}")
            time.sleep(0.5)

            # 종료 연도 & 월 설정
            to_year = Select(driver.find_element(By.XPATH, '//*[@id="selMonthTo"]'))
            to_year.select_by_value(str(year))
            time.sleep(0.5)

            to_month = Select(driver.find_element(By.XPATH, '//*[@id="selDayTo"]'))
            to_month.select_by_value(f"{month:02d}")
            time.sleep(0.5)

            # 검색 버튼 클릭
            search_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="monthPeriodDiv"]/span[2]/input'))
            )
            search_button.click()
            time.sleep(2)  # 데이터 갱신 대기

            # "더보기" 버튼이 있는지 확인 후 클릭
            try:
                more_search = WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.XPATH, '//*[@id="autodanawa_gridC"]/div[3]/article/main/div/div[2]/div[1]/a'))
                )
                more_search.click()
                time.sleep(2)  # 데이터 로딩 대기
            except:
                print("✅ '더보기' 버튼 없음, 계속 진행")

            # 테이블 데이터 추출
            rows = driver.find_elements(By.CSS_SELECTOR, 'table tbody tr')

            rank_count = 0  # 랭킹 카운트 초기화
            for row in rows:
                if rank_count >= 6:  # 랭킹 6위까지만 수집
                    break

                columns = row.find_elements(By.TAG_NAME, 'td')
                if len(columns) < 5:
                    continue  # 잘못된 데이터 스킵

                rank = columns[1].text.strip()  # 랭크
                brand = columns[2].text.strip()  # 브랜드명
                sales = columns[3].text.strip().replace(',', '')  # 판매량
                market_share = columns[4].text.strip().replace('%','')  # 점유율

                if brand in domestic_brands:  # 국산차만 필터링
                    data.append({
                        'Year': year,
                        'Month': month,
                        'Rank': rank,
                        'Brand': brand,
                        'Sales': sales,
                        'Market Share': market_share,
                    })
                    print(f"✔ {year}년 {month}월 - 랭크: {rank}, 브랜드: {brand}, 판매량: {sales}, 점유율: {market_share}")
                    rank_count += 1

except Exception as e:
    print(f"❌ 오류 발생: {e}")

# 크롤링 완료 후 드라이버 종료
driver.quit()

# # 데이터프레임 생성 및 저장
# df = pd.DataFrame(data)
# df.to_csv('domestic_car_sales_data.csv', index=False)
# print("✅ 데이터 저장 완료: domestic_car_sales_data.csv")
```
