```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

import re  # 정규식을 위한 모듈

# ChromeDriver 초기화
driver = webdriver.Chrome()

# 대상 URL 접속
url = 'https://www.kaida.co.kr/ko/statistics/NewRegistList.do'
driver.get(url)

# 페이지 로드 대기
time.sleep(3)  # 필요 시 조정

# 데이터 수집을 위한 반복문 설정 (2020년 1월 ~ 2025년 2월)
data = []
for year in range(15, 26):  # 연도 범위 설정
    for month in range(1, 13):  # 월 범위 설정
        if year == 25 and month > 2:  # 현재 날짜 기준으로 데이터 제한
            break

        # 연도 및 월 선택 (예: 드롭다운 메뉴 조작)
        try:
            year_dropdown = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="dev_setStartYear"]'))  
            )
            year_dropdown.click()
            year_option = driver.find_element(By.XPATH, f'//*[@id="dev_setStartYear"]/option[{year-2}]')
            year_option.click()
            
            

            month_dropdown = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="dev_setStartMonth"]'))  
            )
            month_dropdown.click()
            month_option = driver.find_element(By.XPATH, f'//*[@id="dev_setStartMonth"]/option[{month}]')
            month_option.click()

            # 검색 버튼 클릭
            search_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="mainCenter"]/div/article[1]/div/form/div[2]/div/a'))  # 검색 버튼 ID 확인 필요
            )
            search_button.click()

            time.sleep(3)  

            # 테이블 데이터 추출
            rows = driver.find_elements(By.CSS_SELECTOR, '#statictics_data > div.pop_scroll > table > tbody ')  
            
            for row in rows:
                columns = row.find_elements(By.TAG_NAME, 'tr')  # 각 행의 열(td) 가져오기
                column_count = len(columns)
                for i in range(column_count) :
                    if columns[i].text.strip().split(' ')[0] == 'Land' : # 랜드로버추출
                        brand = 'Land-Rover'
                        sales = columns[i].text.strip().split(' ')[2]
                        ratio = columns[i].text.strip().split(' ')[3]
                    
                    else:

                        brand = columns[i].text.strip().split(' ')[0] # 이외자료 추출
                        sales = columns[i].text.strip().split(' ')[1]
                        ratio = columns[i].text.strip().split(' ')[2]
                   
                        # 데이터 저장
                    data.append({
                            'Year': f"20{year}",
                            'Month': month,
                            'Brand': brand,
                            'Sales': sales,
                            'Ratio': ratio
                        })
    
                    print(f"20{year}년 {month}월 {brand} 판매량: {sales}, 비율: {ratio}")
                

        except Exception as e:
            print(f"Error occurred for {year}-{month}: {e}")
            continue

# 크롤링 완료 후 드라이버 종료
driver.quit()

# 데이터프레임 생성 및 저장
df = pd.DataFrame(data)
df.to_csv('kaida_sales_data.csv', index=False)
print("Data saved to imported_car_data.csv")
```
